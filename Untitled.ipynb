```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns = ['idnumber', 'dob_mm', 'dob_wk', 'bfacil', 'fagecomb','mager41', 'lbo', 'tbo', 'cig_1', 'cig_2', 'cig_3','rf_ncesar', 'dmeth_rec',
              'bwtr14', 'ab_vent6', 'ab_antibio', 'aged']

# bfacil3/Birth Place Recode
# fagecomb/Father's age
# mager41/Mather's age
# rf_ncesar/Number of Previous Cesarean
```

```python
dead = pd.read_csv('d.csv', usecols=columns)
```

```python
live = pd.read_csv('l.csv', usecols=columns)
```

```python
lv_smp = live[live.idnumber.isna()].sample(len(dead))
```

```python
lv_smp = lv_smp.drop(columns=['idnumber', 'aged'])
```
